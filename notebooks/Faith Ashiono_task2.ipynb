```python
#normalize
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
column_names_to_normalize = ['Dur. (ms)','Number of sessions','Total (Bytes)']
X = data_task2[column_names_to_normalize].values
x_scaled = normalizer.fit_transform(X, y =None)
data_temp = pd.DataFrame(x_scaled, columns=column_names_to_normalize, index = data_task2.index)
data_task2[column_names_to_normalize] = data_temp
data_temp.head()
```

```python
#Kmeans cluster
#elbow method

from sklearn.cluster import KMeans
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(data_temp)
    distortions.append(kmeanModel.inertia_)
    
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()    
```

```python
#fitting kmeans to dataset
X = data_temp.values
kmeans = KMeans(n_clusters=3, init= 'k-means++', n_init = 10, max_iter = 300, random_state= 0 )
Y_kmeans = kmeans.fit_predict(X)
# #visualizing clusters
plt.scatter(X[Y_kmeans == 0,0], X[Y_kmeans==0,1 ], s = 100, c='red', label= 'cluster1')
plt.scatter(X[Y_kmeans == 1,0], X[Y_kmeans==1,1 ], s = 100, c='cyan', label= 'cluster2')
plt.scatter(X[Y_kmeans == 2,0], X[Y_kmeans==2,1 ], s = 100, c='green', label= 'cluster3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s =300, c='yellow', label = 'centroids')
plt.title('Clusters')
plt.legend()
plt.show()
cluster_1 = pd.DataFrame(X[Y_kmeans == 0], columns=column_names_to_normalize)
cluster_2 = pd.DataFrame(X[Y_kmeans == 1], columns=column_names_to_normalize)
cluster_3 = pd.DataFrame(X[Y_kmeans == 2], columns=column_names_to_normalize)
```
